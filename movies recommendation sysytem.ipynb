```python
user_ids = rating.userId.unique().tolist()
movie_ids = rating.movieId.unique().tolist()
print("Number of users :{}".format(len(user_ids)))
print("number of Movies:{}".format(len(movie_ids)))
```

```python
movie_data = pd.merge(rating,movies,on = 'movieId')
movie_data.head()
```

```python
movie_data.groupby('title')['rating'].mean().head()
```

```python
movie_data.groupby('title')['rating'].mean().sort_values(ascending =False).head()
```

```python
movie_data.groupby('title')['rating'].count().sort_values(ascending =False).head()
```

```python
#execute the following script to create rating mean count datframe and 
#first add the avareage rating of each movie to this daraframe:
rating_mean_count=pd.DataFrame(movie_data.groupby('title')['rating'].mean())
#add  rating counts as columns
rating_mean_count['rating_counts']=pd.DataFrame(movie_data.groupby('title')['rating'].count())
rating_mean_count.head()
```

```python
movie_data.info()
```

```python
user_movie_rating=movie_data.pivot_table(index='userId',columns='title',values='rating')
user_movie_rating
```

```python
forrest_gump_rating=user_movie_rating['Lord of the Rings: The Return of the King, The (2003)']
forrest_gump_rating.head()
```

```python
movies_like_forest_gump=user_movie_rating.corrwith(forrest_gump_rating)
corr_forest_gump=pd.DataFrame(movies_like_forest_gump,columns=['Correlation'])
corr_forest_gump.dropna(inplace=True)
corr_forest_gump
```

```python
corr_forest_gump.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_forest_gump=corr_forest_gump.join(rating_mean_count['rating_counts'])
corr_forest_gump.head()
```
